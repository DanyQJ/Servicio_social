# Introducción al Machine Learning

El **Machine Learning (Aprendizaje Automático)** es una rama de la inteliengia artificial que nos permite de programar computadoras para que aprendan a partir de los datos. A diferencia de la programación de software tradicional, en la cual desarrollamos reglas analíticas fijas, un algoritmo de ML procesa observaciones y resultados esperados para inferir por sí mismo las reglas matemáticas subyacentes.

## ¿Cómo funciona?

1. **Recolección de datos**: El modelo necesita ejemplos (datos) para aprender.
2. **Entrenamiento**: Se alimenta al algoritmo con datos.
3. **Evaluación**: Se valida el desempeño del modelo con datos no vistos.
4. **Predicción o acción**: El modelo ya entrenado se usa sobre nuevos datos.



# Tipos de Aprendizaje en Machine Learning


El aprendizaje automático (Machine Learning) se divide fundamentalmente en distintas categorías según la naturaleza de los datos de entrenamiento y la forma en que el algoritmo interactúa con ellos para aprender. A continuación, exploraremos los enfoques teóricos más importantes.

## 1. Aprendizaje Supervisado (Supervised Learning)

Para este caso, el modelo se entrena utilizando un **conjunto de datos etiquetado**. Esto significa que para cada ejemplo de entrada en nuestros datos, se proporciona de antemano la salida correcta o el "valor objetivo" (Target). El objetivo del algoritmo es aprender una función de mapeo desde las entradas hasta las salidas, de modo que pueda predecir la etiqueta de nuevos datos nunca antes vistos.

**Características principales:**
* **Datos:** Entradas (Características/Features) $X$ y Salidas (Etiquetas/Labels) $Y$.
* **Objetivo:** Minimizar el error entre la predicción del modelo y el valor real.

**Subcategorías principales:**
* **Clasificación:** La variable de salida es categórica o discreta (ej. predecir si un correo es *Spam* o *No Spam*, o si un tumor es *Maligno* o *Benigno*). Algoritmos comunes: *Regresión Logística, Máquinas de Vectores de Soporte (SVM), Random Forest, Naive Bayes*.
* **Regresión:** La variable de salida es continua o numérica (ej. predecir el precio de una casa en función de sus metros cuadrados, o la temperatura de mañana). Algoritmos comunes: *Regresión Lineal, Regresión Ridge/Lasso*.

## 2. Aprendizaje No Supervisado (Unsupervised Learning)

A diferencia del aprendizaje supervisado, aquí **no existen etiquetas** ni resultados correctos predefinidos. El modelo recibe un conjunto de datos exclusivamente de variables de entrada ($X$) y se deja a su propio criterio para explorar la estructura de la información. 

**Características principales:**
* **Datos:** Únicamente entradas (Features) $X$, sin valores objetivo.
* **Objetivo:** Encontrar patrones ocultos, agrupaciones naturales o simplificar la representación de los datos.

**Subcategorías principales:**
* **Agrupamiento (Clustering):** Agrupa los datos en clústeres basados en la similitud matemática entre ellos. Los elementos del mismo grupo son más similares entre sí que con los de otros grupos (ej. segmentación de clientes). Algoritmos comunes: *K-Means, DBSCAN, Clustering Jerárquico*.
* **Reducción de Dimensionalidad:** Comprime la información reduciendo el número de variables aleatorias a considerar, manteniendo la mayor cantidad de información o varianza posible. Útil para visualización o para acelerar otros modelos. Algoritmos comunes: *Análisis de Componentes Principales (PCA), t-SNE*.
* **Reglas de Asociación:** Descubre relaciones interesantes entre variables en bases de datos grandes (ej. "los clientes que compran pan también suelen comprar leche"). Algoritmos comunes: *Apriori, Eclat*.

## 3. Aprendizaje por Refuerzo (Reinforcement Learning)

Este enfoque está inspirado en la psicología conductista. Aquí no tenemos datos estáticos; en su lugar, tenemos un **agente** que interactúa dinámicamente con un **entorno**. El agente toma acciones y, en respuesta, el entorno le devuelve un nuevo estado y una **recompensa** (positiva o negativa). 

* **Objetivo:** Aprender una política (policy) que maximice la recompensa acumulada a largo plazo. Es el aprendizaje mediante "prueba y error".
* **Aplicaciones clásicas:** Robótica, navegación de vehículos autónomos, y sistemas capaces de jugar videojuegos o juegos de mesa complejos (como AlphaGo o ajedrez).

## 4. Otros tipos de aprendizaje

* **Aprendizaje Semi-supervisado (Semi-supervised Learning):** Utiliza una pequeña cantidad de datos etiquetados y una gran cantidad de datos no etiquetados. Es muy útil cuando extraer los datos es barato, pero etiquetarlos es costoso y requiere expertos humanos (ej. imágenes médicas).
* **Aprendizaje Autosupervisado (Self-supervised Learning):** Una variante donde los datos mismos generan las etiquetas de forma heurística. El modelo retiene una parte de los datos y trata de predecirla utilizando el resto (muy usado hoy en día en Modelos de Lenguaje Grande o LLMs para NLP).

# Flujo de Trabajo Típico en un Proyecto de Machine Learning


El desarrollo de un modelo de Machine Learning no se limita solo a programar algoritmos; es un proceso iterativo y estructurado, frecuentemente guiado por metodologías estándar de la industria como **CRISP-DM** (Cross-Industry Standard Process for Data Mining). A continuación, exploraremos las fases de este ciclo de vida.

## 1. Entendimiento del Problema (Business Understanding)
Antes de escribir una sola línea de código, es imperativo definir claramente el problema que se intenta resolver.
* **Objetivos:** ¿Qué queremos lograr exactamente? (Ej. predecir el abandono de clientes, detectar transacciones fraudulentas, clasificar imágenes médicas).
* **Métricas de Éxito:** Definir cómo se evaluará el impacto del modelo, alineando las métricas técnicas (como la precisión) con los objetivos del proyecto.

## 2. Recolección de Datos (Data Collection)
Consiste en la obtención de la "materia prima" necesaria para que el modelo pueda aprender.
* **Fuentes de Datos:** Extracción de información desde bases de datos (SQL, NoSQL), consumo de APIs, lectura de archivos estructurados (CSV, Parquet) o técnicas como *web scraping*.
* **Evaluación Inicial:** Determinar si la cantidad y calidad de los datos históricos son suficientes para abordar el problema.

## 3. Análisis Exploratorio y Preprocesamiento (EDA)
Históricamente, esta es la fase que consume la mayor parte del tiempo de un Científico de Datos.
* **Limpieza de Datos (Data Cleaning):** Manejo de valores faltantes o nulos (NaN), eliminación de registros duplicados y tratamiento de valores atípicos (*outliers*).
* **Análisis Exploratorio de Datos (EDA):** Uso de estadística descriptiva y visualizaciones (histogramas, diagramas de caja, gráficos de dispersión) para entender la distribución subyacente. Aquí se calculan **correlaciones** (decidiendo rigurosamente entre *Pearson* o *Spearman* según la normalidad y linealidad de los datos).
* **Ingeniería de Características (Feature Engineering):** Creación de nuevas variables predictivas a partir de las existentes, escalado/normalización de variables numéricas y codificación matemática de variables categóricas (ej. *One-Hot Encoding*).

## 4. Selección y Entrenamiento del Modelo
Fase central donde el algoritmo extrae patrones matemáticos de los datos procesados.
* **División del Conjunto de Datos:** Separación estricta de la información en un conjunto de entrenamiento (*Train set*) y un conjunto de prueba (*Test set*) para evitar el sobreajuste (*overfitting*).
* **Selección de Algoritmos:** Elección de las arquitecturas adecuadas (Regresión Lineal, Árboles de Decisión, Random Forest, Redes Neuronales) basadas en la naturaleza del problema (supervisado vs. no supervisado).
* **Entrenamiento (Fit):** Ajuste de los parámetros internos del algoritmo utilizando exclusivamente los datos de entrenamiento.

## 5. Evaluación del Modelo
Comprobación empírica del rendimiento del modelo frente a datos que nunca antes ha visto (el conjunto de prueba).
* **Para Clasificación:** Se evalúa utilizando métricas como *Accuracy* (Exactitud), *Precision*, *Recall*, *F1-Score* y el análisis de la *Matriz de Confusión*.
* **Para Regresión:** Se utilizan métricas de error como el *Error Cuadrático Medio (MSE)*, *Raíz del Error Cuadrático Medio (RMSE)* o el *Coeficiente de Determinación ($R^2$)*.

## 6. Optimización de Hiperparámetros (Model Tuning)
Si el modelo tiene un buen rendimiento base pero requiere mejoras, se procede a ajustar sus hiperparámetros (las "perillas" de configuración que el modelo no aprende por sí solo, como la profundidad máxima de un árbol). 
* Se aplican técnicas de búsqueda exhaustiva o aleatoria (*Grid Search*, *Random Search*) combinadas con **Validación Cruzada** (*Cross-Validation*) para asegurar la robustez del modelo.

## 7. Despliegue (Deployment)
El modelo entrenado y validado sale del entorno de experimentación (Jupyter) para integrarse en un sistema real.
* **Empaquetado:** Exportación del modelo en formatos serializados (como archivos `.pkl`, `.joblib` o `.onnx`) e integración en contenedores Docker.
* **Servicio:** Creación de puntos de enlace (APIs utilizando *FastAPI* o *Flask*) para recibir nuevos datos en formato JSON y devolver predicciones en tiempo real.

## 8. Monitoreo y Mantenimiento
El entorno real es dinámico, por lo que el trabajo continúa tras el despliegue.
* **Monitoreo de Degradación:** Los patrones de los datos cambian con el tiempo (fenómenos conocidos como *Data Drift* o *Concept Drift*), lo que degrada las predicciones.
* **Reentrenamiento Continuo:** Es necesario establecer un sistema que vigile el rendimiento del modelo en producción y desencadene reentrenamientos periódicos con datos frescos para mantener su vigencia.

# Regresión Lineal

La **regresión lineal** es una técnica estadística y de machine learning supervisado que permite modelar la **relación entre una variable dependiente** $y$ y **una o más variables independientes** $x$.



## 1. Objetivo

Encontrar una función lineal que **aproxime** los datos de entrenamiento lo mejor posible.

Para una sola variable:

$$
y = \beta_0 + \beta_1 x + \varepsilon
$$

- $y$: Variable dependiente (respuesta).
- $x$: Variable independiente (predictor).
- $\beta_0$: Intercepto (ordenada al origen).
- $\beta_1$: Pendiente (cuánto cambia $y$ por cada unidad de $x$).
- $\varepsilon$: Término de error aleatorio.


## 2. Geometría de la regresión

La regresión lineal busca una **línea recta** que minimice la **suma de los errores al cuadrado** entre las predicciones y los valores reales.



## 3. Cálculo de los coeficientes (Mínimos Cuadrados)

Dado un conjunto de datos $(x_1, y_1), (x_2, y_2), ..., (x_n, y_n)$, los parámetros se estiman como:

$$
\hat{\beta}_1 = \frac{ \sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y}) }{ \sum_{i=1}^{n} (x_i - \bar{x})^2 }
$$

$$
\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x}
$$

Donde:
- $\bar{x}$: Promedio de las $x_i$
- $\bar{y}$: Promedio de las $y_i$



## 4. Métricas de Evaluación

### Error cuadrático medio (MSE):

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

#### ¿Por qué usar MSE?
- Penaliza **fuertemente los errores grandes** al elevar al cuadrado las diferencias.
- Es **continuo y diferenciable**, lo que lo hace útil para optimización.
- Se interpreta como la **media del error al cuadrado**: cuanto menor sea el MSE, **mejor el ajuste** del modelo.


### Coeficiente de determinación $R^2$:

$$
R^2 = 1 - \frac{ \sum (y_i - \hat{y}_i)^2 }{ \sum (y_i - \bar{y})^2 }
$$

Indica qué tanto del comportamiento de $y$ es explicado por el modelo.

#### ¿Por qué usar $R^2$?
- Mide la **proporción de la varianza explicada** por el modelo respecto a la varianza total.
- Su valor está entre 0 y 1 (o puede ser negativo si el modelo es muy malo).
- $R^2 = 1$ indica **predicción perfecta**, y $R^2 = 0$ indica que el modelo **no explica nada** más que la media.
- Es una medida **relativa del ajuste**, útil para **comparar modelos**.


## 5. Supuestos de la Regresión Lineal

1. **Linealidad** entre las variables.
2. **Independencia** de los errores.
3. **Homoscedasticidad**: varianza constante del error.
4. **Normalidad** del término de error $\varepsilon$.

```python
# Generamos datos sintéticos tipo "PM10 -> PM25"

# Inicializamos el generador de números aleatorios
rng = np.random.default_rng(seed=42)

# 1. Definimos el tamaño de nuestra muestra
# 'n' es el número de observaciones o filas en nuestro dataset.
n = 180

# 2. Generamos la variable independiente (Feature / X)
# Simulamos niveles de PM10 entre 15 y 120 usando una distribución uniforme.
# Estas son las "pistas" que el modelo usará para aprender.
pm10 = rng.uniform(15, 120, n)

# 3. Generamos la variable dependiente (Target / y) basándonos en una fórmula real
# Aquí aplicamos la ecuación de la recta: y = intercepto + (pendiente * x) + error
# 4.5 es el intercepto (beta_0)
# 0.58 es la pendiente (beta_1)
# rng.normal(0, 3.2, n) añade "ruido blanco" para que los datos no sean perfectos.
pm25 = 4.5 + 0.58 * pm10 + rng.normal(0, 3.2, n)

# 4. Creamos el DataFrame (Estructura de datos)
# Organizamos los datos en columnas para que sean fáciles de procesar por algoritmos de ML.
df_pm = pd.DataFrame({
    "PM10": pm10, 
    "PM25": pm25
})

df_pm.head()  # Mostramos las primeras filas para verificar que los datos se han generado correctamente
```

Ahora graficamos con la intención de confirmar que existe una correlación lineal entre nuestros datos.

```python
plt.figure(figsize=(5, 5), dpi=150) # Tamaño y resolución de la figura
sns.scatterplot(x="PM10", y="PM25", data=df_pm) # Gráfico de dispersión para visualizar la relación entre PM10 y PM25
plt.title("Relación sintética entre PM10 y PM2.5") # Título del gráfico para contextualizar lo que estamos viendo
plt.grid(True) # Agrega una cuadrícula para facilitar la lectura de los puntos en el gráfico
plt.show()
```

```python
# Entrenamiento del modelo de regresión lineal
x = df_pm[["PM10"]] # 'x' es un DataFrame de una sola columna (PM10) que representa la variable independiente.
y = df_pm["PM25"] # 'y' es una Serie que representa la variable dependiente.

modelo_lin = LinearRegression() # Creamos una instancia del modelo de regresión lineal.
modelo_lin.fit(x, y) # Entrenamos el modelo usando 'x' como entrada y 'y' como objetivo. El modelo aprende la relación entre PM10 y PM25.

# Predicciones y metricas de error
y_pred = modelo_lin.predict(x) # Usamos el modelo entrenado para predecir los valores de PM25 basándonos en los valores de PM10.

# El MSE mide el promedio de los errores al cuadrado entre los valores reales (y) y las predicciones (y_pred).
# R^2 mide la proporción de la varianza en 'y' que es explicada por 'x' a través del modelo. Un valor cercano a 1 indica un buen ajuste.
# La varianza de 'y' nos da una idea de la dispersión de los valores reales de PM25. Es útil para entender la variabilidad en los datos.
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
var_y = np.var(y)

print(f"Intercepto (beta_0): {modelo_lin.intercept_:.3f}")
print(f"Pendiente (beta_1): {modelo_lin.coef_[0]:.3f}")
print(f"MSE: {mse:.3f}")
print(f"R^2: {r2:.3f}")
print(f"Varianza de y: {var_y:.3f}")
```

```python
plt.figure(figsize=(5, 5), dpi=150) # Tamaño y resolución de la figura
sns.scatterplot(x="PM10", y="PM25", data=df_pm, label="Datos reales") # Gráfico de dispersión para visualizar los datos reales
plt.plot(x, y_pred, color="red", label="Modelo lineal") # Línea de regresión lineal ajustada por el modelo
plt.title("Ajuste de regresión lineal") # Título del gráfico para contextualizar lo que estamos viendo
plt.legend() # Agrega una leyenda para diferenciar entre los datos reales y la línea del modelo
plt.grid(True) # Agrega una cuadrícula para facilitar la lectura de los puntos y la línea del modelo en el gráfico
plt.show()
```
